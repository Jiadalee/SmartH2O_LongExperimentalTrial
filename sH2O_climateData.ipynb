# Data loading

```python
climateData=pd.read_excel('ClimateData_Valencia.xlsx') # Valencia climate data 1950-2018
climateData.keys()
```

```python
# ::: Water use change 

width = 0.35  # the width of the bars

plt.rcParams['figure.figsize'] = [10,7.5]
fig, ax = plt.subplots(2,2)


# ::: Temperature July 
# Add some text for labels, title and custom x-axis tick labels, etc.
ax[0,0].bar(climateData['year'], climateData['tm_mes_Jul'], label='1950-2018 time series', color = 'silver', alpha=0.7, edgecolor = 'white', linewidth=0.3)
ax[0,0].bar(climateData['year'][-4:].values, climateData['tm_mes_Jul'][-4:].values, label='SmartH2O longitudinal study', color = 'coral', alpha=1)
ax[0,0].plot([1949, 2019], [np.mean(climateData['tm_mes_Jul']), np.mean(climateData['tm_mes_Jul'])], ':', color = 'black', label='1950-2018 average')
ax[0,0].legend(fontsize = 8)

ax[0,0].set_ylabel('Average monthly temperature in July [ËšC]')
ax[0,0].set_xlim([1949, 2019])

ax[0,0].set_ylim([20, 28])
x_ticks = np.arange(1950, 2018, 5)
ax[0,0].set_xticks(x_ticks)
ax[0,0].tick_params(axis='x', rotation=90)

fig.tight_layout()
plt.show()

# ::: Temperature August
# Add some text for labels, title and custom x-axis tick labels, etc.
ax[0,1].bar(climateData['year'], climateData['tm_mes_Aug'], label='1950-2018 time series', color = 'silver', alpha=0.7, edgecolor = 'white', linewidth=0.3)
ax[0,1].bar(climateData['year'][-4:].values, climateData['tm_mes_Aug'][-4:].values, label='SmartH2O longitudinal study', color = 'coral', alpha=0.6)
ax[0,1].plot([1949, 2019], [np.mean(climateData['tm_mes_Aug']), np.mean(climateData['tm_mes_Aug'])], ':', color = 'black', label='1950-2018 average')
ax[0,1].legend(fontsize = 8)

ax[0,1].set_ylabel('Average monthly temperature in August [ËšC]')
ax[0,1].set_xlim([1949, 2019])

ax[0,1].set_ylim([20, 28])
x_ticks = np.arange(1950, 2018, 5)
ax[0,1].set_xticks(x_ticks)
ax[0,1].tick_params(axis='x', rotation=90)


# ::: Precipitation July
# Add some text for labels, title and custom x-axis tick labels, etc.
ax[1,0].bar(climateData['year'], climateData['p_Jul (mm)'], label='1950-2018 time series', color = 'silver', alpha=0.7, edgecolor = 'white', linewidth=0.3)
ax[1,0].bar(climateData['year'][-4:].values, climateData['p_Jul (mm)'][-4:].values, label='SmartH2O longitudinal study', color = 'deepskyblue', alpha=0.6)
ax[1,0].plot([1949, 2019], [np.mean(climateData['p_Jul (mm)']), np.mean(climateData['p_Jul (mm)'])], ':', color = 'black', label='1950-2018 average')
ax[1,0].legend(fontsize = 8)

ax[1,0].set_ylabel('Cumulative monthly precipitation in July [mm]')
ax[1,0].set_xlim([1949, 2019])

ax[1,1].set_ylim([0, 50])
x_ticks = np.arange(1950, 2018, 5)
ax[1,0].set_xticks(x_ticks)
ax[1,0].tick_params(axis='x', rotation=90)
fig.tight_layout()
plt.show()

# ::: Precipitation August
# Add some text for labels, title and custom x-axis tick labels, etc.
ax[1,1].bar(climateData['year'], climateData['p_Aug (mm)'], label='1950-2018 time series', color = 'silver', alpha=0.7, edgecolor = 'white', linewidth=0.3)
ax[1,1].bar(climateData['year'][-4:].values, climateData['p_Aug (mm)'][-4:].values, label='SmartH2O longitudinal study', color = 'deepskyblue', alpha=0.6)
ax[1,1].plot([1949, 2019], [np.mean(climateData['p_Aug (mm)']), np.mean(climateData['p_Aug (mm)'])], ':', color = 'black', label='1950-2018 average')
ax[1,1].legend(fontsize = 8)

ax[1,1].set_ylabel('Cumulative monthly precipitation in August [mm]')
ax[1,1].set_xlim([1949, 2019])

ax[1,1].set_ylim([0, 120])
x_ticks = np.arange(1950, 2018, 5)
ax[1,1].set_xticks(x_ticks)
ax[1,1].tick_params(axis='x', rotation=90)
fig.tight_layout()
plt.show()

fig.savefig("PaperFigs/S8_TemperaturePrecipitation.pdf", bbox_inches='tight')


```
